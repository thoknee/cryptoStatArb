```python
from binance.client import Client as bnb_client
from datetime import datetime
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
client = bnb_client(tld='US')

def get_data(symbol,freq,start_ts = '2025-05-01',end_ts='2025-07-15'):
    data = client.get_historical_klines(symbol,freq,start_ts,end_ts)
    columns = ['open_time','open','high','low','close','volume','close_time','quote_volume',
    'num_trades','taker_base_volume','taker_quote_volume','ignore']

    data = pd.DataFrame(data,columns = columns)
    
    # Convert from POSIX timestamp (number of millisecond since jan 1, 1970)
    data['open_time'] = data['open_time'].map(lambda x: datetime.utcfromtimestamp(x/1000))
    data['close_time'] = data['close_time'].map(lambda x: datetime.utcfromtimestamp(x/1000))
    return data
```

```python
universe = ['BTCUSDT', 'ETHUSDT', 'BNBUSDT', 'SOLUSDT', 'ADAUSDT', 
            'XRPUSDT', 'MATICUSDT', 'DOGEUSDT']
freq = '1h'


px = {}
for x in universe:
    data = get_data(x,freq)
    px[x] = data.set_index('open_time')['close']

px = pd.DataFrame(px).astype(float)
px = px.reindex(pd.date_range(px.index[0],px.index[-1],freq=freq))
ret = px.pct_change()
```

```python
returns = ret.stack().rename("return").reset_index()
returns['hour'] = returns['level_0'].dt.hour
returns['month'] = returns['level_0'].dt.month
returns['day'] = returns['level_0'].dt.day
returns['weekday'] = returns['level_0'].dt.dayofweek  # 0 = Monday
returns['hour'] = returns['level_0'].dt.hour
returns
```

```python
monthly_avg = returns.groupby('hour')['return'].mean()
monthly_avg.plot(kind='bar', title="Average Return by Hour", ylabel='Avg Return')
```

```python
goodDays = [0,2,4,5]
days = returns[returns['weekday'].isin(goodDays)]
hours = days[(days['hour'] == 1) | (days['hour'] == 23)]

```

```python
hours['position'] = hours['hour'].apply(lambda x: 1 if x == 1 else -1)
hours
```

```python
strat_returns = hours['return'] * hours['position']
strat_returns.sum()
```
